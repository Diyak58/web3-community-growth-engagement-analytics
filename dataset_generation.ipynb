```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# make results reproducible
np.random.seed(42)
random.seed(42)

print("Libraries loaded successfully!")
```

```python
sources = [
    "Twitter",
    "Discord Invite",
    "Referral",
    "Event",
    "Ambassador Program"
]

countries = [
    "India",
    "USA",
    "UK",
    "Germany",
    "Singapore",
    "Canada",
    "Brazil",
    "Nigeria",
    "Philippines",
    "Australia"
]

print("Categories created successfully!")
```

```python
num_users = 5000

start_date = datetime(2024, 1, 1)
end_date = datetime(2025, 12, 31)

users = pd.DataFrame({
    "user_id": [f"U{i:04d}" for i in range(1, num_users + 1)],
    "join_date": [
        start_date + timedelta(
            days=random.randint(0, (end_date - start_date).days)
        )
        for _ in range(num_users)
    ],
    "source": np.random.choice(
        sources,
        size=num_users,
        p=[0.35, 0.20, 0.15, 0.15, 0.15]
    ),
    "country": np.random.choice(countries, size=num_users),
    "wallet_connected": np.random.choice(
        ["Yes", "No"],
        size=num_users,
        p=[0.7, 0.3]
    )
})

users.head()
```

```python
users.shape
```

```python
users.to_csv("../Data/users.csv", index=False)

print("users.csv saved successfully!")
```

```python
activity_types = [
    "Message",
    "AMA Join",
    "Event Join",
    "Quest Complete",
    "Referral"
]

print("Activity types created successfully!")
```

```python
activities = []

print("Activity list created!")
```

```python
num_activities = 25000

for i in range(1, num_activities + 1):

    user = users.sample(1).iloc[0]

    activity_date = user["join_date"] + timedelta(
        days=random.randint(0, 180)
    )

    activities.append([
        f"A{i:05d}",
        user["user_id"],
        activity_date,
        random.choice(activity_types),
        random.randint(1, 50)
    ])

print("Activities generated successfully!")
```

```python
activities_df = pd.DataFrame(
    activities,
    columns=[
        "activity_id",
        "user_id",
        "activity_date",
        "activity_type",
        "points_earned"
    ]
)

activities_df.head()
```

```python
activities_df.shape
```

```python
activities_df.to_csv("../Data/activities.csv", index=False)

print("activities.csv saved successfully!")
```

```python
campaign_names = [
    "Twitter Growth Campaign",
    "Airdrop Campaign",
    "Community Quest Program",
    "Referral Drive",
    "Ambassador Recruitment",
    "Discord Expansion",
    "Telegram Growth Push",
    "NFT Giveaway Campaign",
    "AMA Engagement Series",
    "Web3 Education Program",
    "Hackathon Promotion",
    "Partnership Awareness Campaign"
]

print("Campaign names created successfully!")
```

```python
campaigns = pd.DataFrame({
    "campaign_id": [f"C{i:03d}" for i in range(1, 13)],
    "campaign_name": campaign_names,
    "start_date": pd.date_range(
        start="2024-01-01",
        periods=12,
        freq="MS"
    ),
    "spend_usd": np.random.randint(
        2000,
        10000,
        size=12
    )
})

campaigns["end_date"] = campaigns["start_date"] + pd.Timedelta(days=29)

campaigns.head()
```

```python
campaigns.to_csv("../Data/campaigns.csv", index=False)

print("campaigns.csv saved successfully!")
```

```python
user_campaigns = pd.DataFrame({
    "user_id": np.random.choice(
        users["user_id"],
        size=5000,
        replace=False
    ),
    "campaign_id": np.random.choice(
        campaigns["campaign_id"],
        size=5000
    )
})

user_campaigns.head()
```

```python
user_campaigns.to_csv("../Data/user_campaigns.csv", index=False)

print("user_campaigns.csv saved successfully!")
```
